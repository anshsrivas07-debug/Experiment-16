```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={
    'Days':['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],
    'Study_Hours':[2,3,2.5,4,3.5,1,1.5],
    'Marks':[50,60,55,70,85,90,80],
    'Attendance':[60,70,80,75,70,80,86],
    'Sleep_Hours':[6,5,4,8,9,4,6],
    'Assignments_Completed':[1,2,1,3,2,3,2]
}

df=pd.DataFrame(data)
print(df)
```

```python
#1 Line Chart
plt.plot(df['Study_Hours'], df['Marks'], marker='*')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
# Advanced Line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Days'], df['Study_Hours'], label='Study Hours', marker='o',color='red')
plt.plot(df['Days'], df['Marks'], label='Marks', marker='*',color='blue')
plt.title("Study Hours vs Marks Trends")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()
```

```python
# Bar chart(comparison)
plt.figure(figsize=(10, 6))
plt.bar(df['Days'], df['Marks'], label='Study Hours',color='red')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.legend()
plt.show()
```

```python
# ADvanced Bar Graph
plt.figure(figsize=(10, 6))

bars = plt.bar(df['Days'], df['Marks'], color='red', label='Marks')

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y),
             ha='center', va='bottom')

plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Histogram
data={
    'Marks':[50,60,55,70,85,90,80,56,58,45,45,35,15,97,54,10,65,62,76,43]
}
df=pd.DataFrame(data)
plt.hist(df['Marks'],bins=20,color='red',edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Distribution')
plt.show()
```

```python
# Scatter plot
plt.scatter(df['Study_Hours'], df['Marks'],color='red',marker='*')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Study Hours vs Marks')
plt.show()
```

```python
# Conditional Scatter plot
#example category
df['Result']=["Fail","Pass","Fail","Pass","Fail","Pass","Pass"]
color=["red" if x=="Fail" else "green" for x in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'],c=color,marker="*")
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Study Hours vs Marks')
plt.show()
```

SAME GRAPHS USING SEABORN
